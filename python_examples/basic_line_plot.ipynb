```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# set up directory for saving plots
plot_dir = 'example_figures/'
```

### Example line plot

```python
# make up some data
x_val = np.linspace(1,20,20)
y_val = np.concatenate([np.linspace(1,10,10), np.linspace(10,300,10)])

# create a figure 
plt.figure(figsize=(6,5))

# plot our data
plt.plot(x_val, y_val, label='Made up data')

# add some axis labels (w/ units!)
plt.xlabel('Days (d)')
plt.ylabel('[O$_2$] ($\mu$mol kg$^{-1}$)')
plt.title('Example plot of a single line')

plt.legend()

plot_filename = 'single_line_example'

plt.savefig(plot_dir + plot_filename + '.png', dpi=300)
```

### Example "for loop"

```python
len(x_val)
```

```python
# Let's say we want to do some calculation for every day that uses the value from the day before
# we can start with an initial value (t_out[0]) and save subsequent values in t_out

t_out = np.full(len(x_val), np.nan)
t_out[0] = 2

for d in range(1, len(x_val)):
    t_out[d] = t_out[d-1]+ t_out[d-1]/10

# create a figure 
plt.figure(figsize=(6,5))

# plot our data
plt.plot(x_val, t_out, label='example data')

# Add another line  
plt.plot(x_val, t_out/2, label='example data / 2')

# add some axis labels (w/ units!)
plt.xlabel('Days (d)')
plt.ylabel('Some other concentration ($\mu$mol kg$^{-1}$)')
plt.title('Example plot of a two line-plot')

# add a legend
plt.legend()

plot_filename = 'two_line_example'

plt.savefig(plot_dir + plot_filename + '.png', dpi=300)
```
